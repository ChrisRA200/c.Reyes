```python
import pandas as pd
```

```python
fName='data/msft.csv'
df=pd.read_csv(fName)
```

```python
df.describe().T
```

```python
df.head()
```

```python
df=df[['Date','Open','High','Low','Adj Close','Volume']]
df.columns=['Date','Open','High','Low','Close','Volume']
df=df[['Date','Open','High','Low','Volume','Close']]
df.head()
```

```python
df['Mes']=df['Date'].apply(lambda x: int(x[5:7]))
df['Retorno']=df['Close'].pct_change()
df.head()
```

```python
df=df.dropna()
df.head()
```

```python
df['Y']=df['Retorno'].apply(lambda x: 1 if x>0 else 0)
df.head()
```

```python
X=df[['Open','High','Low','Volume','Mes']]
```

```python
def normalize(dataIn):
    return (dataIn-dataIn.min())/(dataIn.max() - dataIn.min())
```

```python
Xnorm=normalize(X)
```

```python
Xnorm.head()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
y=df['Y']
X_train, X_test, y_train, y_test = train_test_split(Xnorm, y, test_size=0.2, random_state=0)
```

```python
gnb = GaussianNB()
```

```python
model = gnb.fit(X_train, y_train)
```

```python
predicciones=model.predict(X_test)
```
